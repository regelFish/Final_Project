```python
import pandas as pd
import glob
import os

output_dir = "mbtace_parquet_files"

# 1. Get a list of all parquet files generated in the previous step
all_files = glob.glob(os.path.join(output_dir, "*.parquet"))

# 2. Use a list comprehension to read all files into a list of DataFrames
# This is faster than appending to a single DataFrame in a loop
print(f"Loading {len(all_files)} files into memory...")
df_list = [pd.read_parquet(file) for file in all_files]

# 3. Concatenate all DataFrames at once
final_df = pd.concat(df_list, ignore_index=True)

print(f"Successfully combined! Total rows: {len(final_df)}")
```
