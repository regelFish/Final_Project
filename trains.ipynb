```python
import polars as pl

pl.enable_string_cache()

# 1. Prep Schedule (Unique)
schedule_lazy = (
    pl.scan_parquet("mbtace_parquet_files/*.parquet")
    .filter(pl.col("route_id").str.starts_with("Green"))
    .select([
        pl.col("trip_id").cast(pl.Categorical),
        pl.col("stop_id").cast(pl.Categorical),
        (pl.col("arrival_time").str.split(":").list.get(0).cast(pl.Int64) * 3600 +
         pl.col("arrival_time").str.split(":").list.get(1).cast(pl.Int64) * 60 +
         pl.col("arrival_time").str.split(":").list.get(2).cast(pl.Int64)
        ).mul(1000).cast(pl.Duration("ms")).alias("sch_dur")
    ])
    .unique(subset=["trip_id", "stop_id"]) 
)

# 2. Prep Actuals (Deduplicated)
actual_lazy = (
    pl.scan_parquet("joined_output.parquet")
    .filter(pl.col("route_id").str.starts_with("Green"))
    .select([
        pl.col("trip_id").cast(pl.Categorical),
        pl.col("to_stop_id").cast(pl.Categorical).alias("stop_id"),
        # Get the max/last arrival time if there are multiple pings for one stop
        (pl.col("to_stop_arrival_datetime").dt.time().cast(pl.Duration("ms")))
        .alias("act_dur"),
        pl.col("precipitation"),
        pl.col("route_id"),
        pl.col("to_stop_arrival_datetime")
    ])
    # THIS REMOVES THE IDENTICAL ROWS SEEN IN YOUR IMAGE
    .unique(subset=["trip_id", "stop_id", "act_dur"]) 
)

# 3. Join and Sink
(actual_lazy.join(
    schedule_lazy,
    on=["trip_id", "stop_id"],
    how="inner"
)
.with_columns(
    (pl.col("act_dur") - pl.col("sch_dur")).alias("delay")
)
.sink_parquet("final_delays_cleaned.parquet")
)

print("Deduplicated join complete!")
```

```python
stops_info = (
    pl.read_csv("MBTA_Rapid_Transit_Stop_Distances.csv")
    .select([
        pl.col("stop_id").cast(pl.String),
        pl.col("stop_name")
    ])
    .unique(subset=["stop_id"]) # This is the missing piece
)

# 2. Join them using Lazy mode
final_result = (
    pl.scan_parquet("final_delays_cleaned.parquet")
    # Cast Parquet ID to string just to be 100% safe for the join
    .with_columns(pl.col("stop_id").cast(pl.String))
    .join(stops_info.lazy(), on="stop_id", how="left")
)
final_result.sink_parquet("final_delays_with_names.parquet")

print("Final version saved to final_delays_with_names.parquet")
```
